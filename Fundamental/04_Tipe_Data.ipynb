# **Tipe Data Primitif**

### Numbers

```python
x = 6
print(type(x))

x = 6.0
print(type(x))

x = 1+2j
print(type(x))
```

```python
var = 10
print(var)
print(id(var))

var = 11
print(var)
print(id(var))
```

### Boolean

```python
x = True
print(type(x))

x = False
print(type(x))
```

### String

```python
x = 'Dicoding'
print(type(x))
```

```python
multi_line = """Halo!
Kapan terakhir kali kita bertemu?
Kita bertemu hari Jum’at yang lalu."""

print(multi_line)
```

```python
x = 'Dicoding'
print(x[0])
```

```python
x = 'Dicoding'
print(x[2:])
```

```python
# Formatted String

name = "Nugie Saputra"
print(f"Hello, nama saya {name}")
```

```python
# %-formating
name = "Perseus Evans"
print("Nama saya %s" % (name))
```

```python
name = "Arsha"
print("Nama saya {}".format(name))
```

# **Tipe Data Collection**

### List

```python
x = [1, 2.2, 'Dicoding']
print(type(x))
```

```python
x = [1, 'Dicoding', True, 1.0]

print(x[2])
```

```python
# List Python bersifat mutable yang artinya nilai di dalamnya dapat diubah.

x = [1, 2.2, 'Dicoding']
x[0] = 'Indonesia'
print(x)
```

```python
x = ["laptop", "monitor", "mouse", "mousepad", "keyboard", "webcam", "microphone"]

print(x[0])
print(x[2])
print(x[-1])
print(x[-3])
```

```python
# sequence[start:stop:step]

x = ["laptop", "monitor", "mouse", "mousepad", "keyboard", "webcam", "microphone"]

print(x[0:5:2])
print(x[1:])
print(x[:3])
```

### Tuple

```python
x = (1, "Dicoding", 1+3j)
print(type(x))
```

```python
x = (5, 'program', 1+3j)
print(x[1])
print(x[0:3])
```

```python
# Tuple bersifat immutable yang artinya tidak dapat diubah.

x = (5, 'program', 1+3j)
x[1] = 'Dicoding'
```

### Set

```python
x = {1, 2, 7, 2, 3, 13, 3}
print(x)
print(type(x))
```

```python
x = {1,2,7,2,3,13,3}
print(x[0])
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

union = set1.union(set2)
print("Union:", union)

intersection = set1.intersection(set2)
print("Intersection:", intersection)
```

### Dictionary

```python
x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }

print(type(x))
```

```python
x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }

print(x[0])
```

```python
x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }

print(x ['name'])
```

```python
# Menambah Data pada Dictionary

x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }
x ['Job'] = "Web Developer"

print(x)
```

```python
# Menghapus Data pada Dictionary

x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }
del x['isMarried']

print(x)
```

```python
# Mengubah Data pada Dictionary

x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }
x ['name'] = "Dicoding"

print(x)
```

# **Konversi antara Tipe Data**

```python
# Konversi Integer ke Float
print(float(5))
```

```python
# Konversi Float ke Integer
print(int(5.6))
print(int(-5.5))
```

```python
# Konversi dari-dan-ke String
print(int("25"))
print(str(25))
print(float("25"))
print(str(25.6))
```

```python
# Konversi Kumpulan Data
print(set([1,2,3]))
print(tuple({5,6,7}))
print(list('hello'))
```

```python
# Konversi ke Dictionary
print(dict([[1,2],[3,4]]))
```

```python
print(dict([(3,26),(4,44)]))
```
