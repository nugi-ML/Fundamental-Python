# Mengubah Huruf Besar/Kecil

### `upper()`

```python
kata = 'indonesia'
kata = kata.upper()
print(kata)
```

### `lower()`

```python
kata = 'INDONESIA'
kata = kata.lower()
print(kata)
```

# Awalan dan Akhiran

### `rstrip()`

```python
print('Indonesia        '.rstrip())
```

### `lstrip()`

```python
print('       Indonesia'.lstrip())
```

### `strip()`

```python
print('   Indonesia   '.strip())
```

```python
kata = 'CodeCodeIndonesiaCodeCode'
print(kata.strip('Code'))
```

### `startswith()`

```python
print('Dicoding Indonesia'.startswith('Dicoding'))
```

### `endswith()`

```python
print('Dicoding Indonesia'.endswith('Dicoding'))
```

# Memisah dan Menggabung String

### `join()`

```python
print(' '.join(['Dicoding', 'Indonesia', '!']))
```

```python
print('123'.join(['Dicoding', 'Indonesia', '!']))
```

### `split()`

```python
print('Dicoding Indonesia !'.split())
```

```python
print('''
Aku adalah Manusia,
Aku menyukai Wanita
Aku tinggal di bumi
'''.split('\n'))
```

# Mengganti Elemen String

### `replace()`

```python
string = 'Ayo belajar Bahasa Python'
print(string.replace('Bahasa', 'Pemrograman'))
```

# Pengecekan String

### `isupper()`

```python
kata = 'DICODING'
print(kata.isupper())
```

### `islower()`

```python
kata = 'dicoding'
print(kata.islower())
```

### `isalpha()`

```python
kata = 'indonesia'
print(kata.isalpha())
```

```python
kata = '1ndonesia'
print(kata.isalpha())
```

### `isalnum()`

```python
kata = '1ndonesia'
print(kata.isalnum())
```

### `isdecimal()`

```python
print('123'.isdecimal())
```

### `isspace()`

```python
print('         '.isspace())
```

### `istitle()`

```python
print('Tanah Air'.istitle())
```

# Formatting pada String

### `zfill()`

```python
teks = 'Code'
tambah_number = teks.zfill(5)
print(tambah_number)
```

### `rjust()`

```python
print('Indonesia'.rjust(20))
```

```python
print('Indonesia'.rjust(20, '!'))
```

### `ljust()`

```python
print('Indonesia'.ljust(20, '*'))
```

### `center()`

```python
print('Indonesia'.center(11, '-'))
```
