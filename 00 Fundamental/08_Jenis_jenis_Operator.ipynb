## Operator Aritmetika

```python
x = 11
y = 5

print(x + y)
print(x - y)
print(x * y)
print(x // y)
print(x / y)
print(x % y)
print(x ** y)
```

## Operator Relasional


```python
x = 5
y = 10

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

```python
x = "Dicoding"
y = "Indonesia"

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

## Operator Logika

### Operator End

```python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

### Operator OR

```python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

### Operator NOT

```python
print(not True)
print(not False)
```

## Operator Assignment


```python
# +=
x = 11
x += 5
print(x)

# -=
x = 11
x -= 5
print(x)

# *=
x = 11
x *= 5
print(x)

# /=
x = 11
x /= 5
print(x)

#%=
x = 11
x%= 5
print(x)
```
